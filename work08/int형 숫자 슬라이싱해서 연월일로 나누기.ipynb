```
# 'date' 열의 int형 숫자를 나눠서 'year', 'month', 'day' 열으로 추가하기
# ex) 202101150000000 ---> 'year' = 2021, 'month' = 01, 'day' = 15
# 영필씨 조 스터디에서 배웠습니다.

import pandas as pd

date = {'date': range(202101150000000,202101150000010)}

df = pd.DataFrame(date)
df
```

```
# 'date' 열의 타입은 int64
df.dtypes
```

```
# .slice 메소드를 사용하기 위해서, .astype 메소드를 통해 'date' 열의 타입을 int64에서 str으로 변경
df['date_s'] = df['date'].astype('str')
df.dtypes
```

```
df.head(3)
```

```
# .apply 메소드는 데이터프레임 메소드이므로 대괄호가 2개이어야 함
# .apply 메소드에는 데이터프레임의 각 열이 시리즈 형태로 인자로 전달됨
# slice는 Series.str.slice(start=None, stop=None, step=None)의 형태로 쓰임
df['year'] = df[['date_s']].apply(lambda x: x.str.slice(start=0, stop=4))
df['month'] = df[['date_s']].apply(lambda x: x.str.slice(start=4, stop=6))
df['day'] = df[['date_s']].apply(lambda x: x.str.slice(start=6, stop=8))
df.head(3)
```

```
# 필요없는 열 제거
df.drop(['date', 'year', 'month', 'day'], axis=1, inplace=True)
```

```
# 대괄호 2개 : dataframe 형태
df[['date_s']].head(3)
```

```
# 대괄호 1개 : series 형태
df['date_s'].head(3)
```

```
# to_datetime() 함수로 int형 데이터를 datetime64 자료형으로 변환하기
df['new_date'] = df[['date_s']].apply(lambda x: x.str.slice(start=0, stop=8))
df['new_date_s'] = pd.to_datetime(df['new_date'])
df.head(3)
```
