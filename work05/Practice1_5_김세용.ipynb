```python
import pandas as pd
import sys
input_file = '../work03/dataset/csv_exam.csv'
output_file = '../work03/dataset/output5.csv'

# csv 파일을 읽어서 데이터 프레임으로 저장
df = pd.read_csv(input_file)

print(df)
# print(df['english'])
# 데이터 프레임에서 enlgish 필드(실수값)를 스트링으로 바꾼 후 $ 문자 제거
# 다시 float로 변환하여 저장
df['english'] = df['english'].str.strip('$').astype(float)
print(df['english'])

# class 필드에 1반이 포함되거나 english가 80보다 큰 데이터만
# 선택하여 새로운 데이터프레임에 저장
# loc[행 필터링 조건 : 열 필터링 조건]
print(df.loc[(df['class'] == 1) | (df['english'] > 80), :])
result = df.loc[(df['class'] == 1) | (df['english'] > 80), :]

# csv에 결과 저장
# index를 내보내고 싶지 않은 경우 index=False
result.to_csv(output_file, index=False)
```
