```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10, 4),
                 columns=['A','B','C','D'],
                 index=np.arange(0,100,10))
df
```

```python
# 데이터프레임의 인덱스(x축)와 값(y축)으로 그래프 출력
# 범례
df.plot()
```

```python
# 매직 명령어
# 그래프를 조작할 수 있다.
# 그래프 제목 오른쪽의 stop interaction 버튼을 누르기 전까지 그래프의 수정이 가능함
%matplotlib nbagg

# 생성된 이후에는 조작할 수 없는 옵션
# %matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# 그래프를 그린 후에 제목 우측의 아이콘을 눌러야 그래프가 마무리된다.
# 그렇지 않으면 이전에 출력된 그래프에 내용이 추가됨
# B 컬럼만 선택하여 그래프를 출력함
df["B"].plot()
```

```python
# 그래프를 그린 후에 제목 우측의 아이콘을 눌러야 그래프가 마무리된다.
# 그렇지 않으면 이전에 출력된 그래프에 내용이 추가됨
# C 컬럼만 선택하여 그래프를 출력함
df["C"].plot()
```

```python
# 그래프를 그린 후에 제목 우측의 아이콘을 눌러야 그래프가 마무리된다.
# 그렇지 않으면 이전에 출력된 그래프에 내용이 추가됨
# D 컬럼만 선택하여 그래프를 출력함
df["D"].plot()
```

```python
# 막대 그래프
# list("abcd...") a,b,c,d 식으로 리스트로 만들어짐
# 시리즈 값을 막대 그래프로 출력
s2 = pd.Series(np.random.rand(16), index=list("abcdefghijklmnop"))
s2
```

```python
# 수직 막대 그래프
s2.plot(kind='bar')
```

```python
# 수평 막대 그래프
s2.plot(kind='barh')
```

```python
# 데이터프레임
df2 = pd.DataFrame(np.random.rand(6, 4),
    index=['one', 'two', 'three', 'four', 'five', 'six'],
    columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))
df2
```
