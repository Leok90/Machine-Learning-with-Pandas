```python
import pandas as pd
from scipy import stats 

# read_csv() 함수로 df 생성
df = pd.read_csv('../dataset/mpg/auto-mpg.csv', header=None)

# 열이름을 지정
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower',
             'weight', 'acceleration', 'model_year', 'origin', 'name']

print("===> df.mpg.count() : ", df.mpg.count())
print("===> df.mpg.mean() : ", df.mpg.mean())
print("===> df.mpg.std() : ", df.mpg.std())
print("===> df.mpg.min() : ", df.mpg.min())
print("===> df.mpg.max() : ", df.mpg.max())
print(stats.scoreatpercentile(df.mpg, 25))
print(stats.scoreatpercentile(df.mpg, 50))
print(stats.scoreatpercentile(df.mpg, 75))
```
