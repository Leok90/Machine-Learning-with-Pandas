```python
# 5.2_dropna.py
# 라이브러리 불러오기
import seaborn as sns

# titanic 데이터셋 가져오기
df = sns.load_dataset('titanic', engine='python')
print(df.shape, '\n')

# for 반복문으로 각 열의 NaN 개수 계산하기
missing_df = df.isnull()
print(missing_df, '\n')
print(missing_df.columns, "\n===>", missing_df.columns.size)
```

```python
# notnull() 메서드로 누락 데이터 찾기
print(df.head().notnull())
```
